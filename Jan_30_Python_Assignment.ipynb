### Q1 Write a programm to accept percentage from the user and display the grade according to the following criteria


```python
marks = float(input("Enter you percentage:"))
if marks > 90 :
    grade = 'A'

elif marks > 80 and marks <=90 :
    grade = 'B'

elif marks >= 60 and marks <= 80 :
    grade = 'C'

elif marks < 60 :
    grade = 'D'
    
print(f"You got {grade}")
```

```python
cost = float(input("Enter your cost price of a bike (Rs) : "))

if cost > 100000 :
    tax=0.15
elif cost > 50000 and cost <= 100000 :
    tax=0.1
elif cost <= 50000 :
    tax=0.05

road_tax = round(cost * tax,2)
print(f"You need to pay {road_tax} as your road tax for bike")
```

### Q4 How many times a given number can divided by 3 before it is less than or equal to 10.

```python
num = int(input("Enter the number"))
count = 0
for i in range(num-3,10,-3):
    count+=1
print(f"Number can be divided by 3 = {count} times")
```

```python
# Alternate solution for Q4
num = int(input("Enter the number"))
count = 0
#In "if" we are checking if number is already less or equal to 10. If it is then count is zero
if num<=10 :
    count = 0
else:
# In else, as we dont want to number be less or equal to 10, so substracting 11 from number and then divide by 3 which will
# give desired count
    count = int((num-11)/3)
print(f"Number can be divided by 3 = {count} times")
```

### Q5 Why and when to use while loop in python give a detailed description with example.

1. When we don't know how many times loop should run that time we can make use of WHILE loop.
2. We can specify condition & based on that condition while loop will end.
3. While loop is used to execute block of code multiple times until mentioned condition true and once condition is becomes false loop ends

eg. Lets take example in Q4.
In Q4 we dont know how many times we have to divide by 3 but there is certain condition. We have to divide by 3 only till that condition is true. i.e number > 10

So, lets see below code for while loop -

```python
num = int(input("Enter the number"))
count = 0
# here we have specified the condition for while loop num-3 > 10
# here we are comparing num-3 as we have to stop befor number becomes 10 or less than 10
# so befor dividing we are checking if we divide number then does number fall under or equal to 10
while num-3 > 10 :
    num = num - 3 
    count+=1
# loop will ends after condition becomes false
print(f"number can be divided by {count} times")
```

### Q6 Use nested while loop to print 3 different pattern.

```python
# Right Angled Triangle

num = 8
i=num

while i>0:
    j = num-(i-1)
    while j:
        print('*',end='')
        j-=1
    print('')
    i-=1
```

```python
# Rectangle 
height = 5
length = 8

i=height
j=length

while i:
    j=length
    while j:
        print('* ' ,end='')
        j-=1
    print('')
    i-=1
```

```python
# Trainlge
num = 8
i=num

while i > 0:
    j = 0
    k = num - (i-1)
    while j <= i/2 :
        print(' ',end='')
        if j==i/2:
            while k:  
                print('*',end='')
                k-=1
        j+=1
    print('')
    i-=1
```
