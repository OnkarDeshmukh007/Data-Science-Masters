Multithreading is the process of executing the multiple threads in a processor.

To achive multithreading in python we use a mocule named "threading"

threading module is used to be able to do multithreading. To execute multiple threads at a time

1. activeCount() - instead of this active_count() is now used as activeCount() function is deprecated
    - this returns count of thread objects which are currently alive based on list returned by enumerate method
2. currentThread() - this is derecated instead use current_thread()
    - this returns the current thread object.
3. enumerate() - 
    - this returns the list of all current threads which are alive.

1. run() is a method from Tread class from threading module - Which invokes the callable object passed to constructor of its class. we can override the run method by inheriting the class.

2. start() - it starts the thread, also invokes the run() method by itself. 

3. join() - this method blocks the calling thread until the thread whose join method is called terminates

4. isAlive() - tells if thread is alive or not (True/Flase)

```python
def square(n):
    lst=[]
    for i in range(1,n):
        lst.append(i**2)
    print('Squares of {} is {}'.format(range(1,n),lst))

def cube(n):
    lst=[]
    for i in range(1,n):
        lst.append(i**3)
    print('Cubes of {} is {}'.format(range(1,n),lst))
```

```python
t1 = threading.Thread(target=square, args = (5,))
t2 = threading.Thread(target=cube, args = (5,))
```

```python
t1.start()
```

```python
t2.start()
```

```python
t1.join()
```

```python
t2.join()
```

- Advantages -
    1. As threads shares resource which enables the applications to have multiple thread in single address space
    2. Using multithreading applications/codes becomes more responsive (multiple threads will do multiple different actions)
    3. Time is saved by executing multiple threads at same time
    4. In case of multi processor architecutre, we can run threads simultanously in multiple cores
    5. Effective utilization of resourcess

- Disadvantages -
    1. There is context switching overhead
    2. It is comples to write,understand and to debug
    3. More chances of deadlock
    4. Chancess of race conditions

- DeadLocks -
    - Deadlocks is the scenario where threads are waiting for some condition which never happens
    - Like a thread is waiting for another thread to release resourcess and another thread waiting for 1st thread
    - There colud be multiple reasons why this happens one of them mentioned above. Others cloud be thread fails to realease resourses/locks on resources
    - Thread is waiting on itself

- Race -
    - Is the failure where programm behavior depends on the order of execution of threads
    - Consider two threads writing in same variable. and as we know thread context switch happens anytime while execution. So if thread writes value as 100 and 2nd thread then writes as 200, final value of variable depends on which thread executed at last i.e programm behavior is determined by order in which threads executed. and this is a race condition.
