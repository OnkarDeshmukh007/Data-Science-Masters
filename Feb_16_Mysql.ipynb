### Q1. What is a database? Differentiate between SQL and NoSQL databases.

Database - Is a collection of data stored on computer systems in organised manner

- SQL vs NoSQL Databses - 
    1. SQL databases are Relational Databases and NoSQLs are Non-relational 
    2. SQL databases stores data in form of tables (rows & columns), NoSQL databases stores data in form of documents (json formatt) or Key-value pairs or Graphs or Wide-columnar
    3. SQL databases follows ACID properties, NoSQL follows the CAP theorem
    4. In SQL databases, tables have fixed schema i.e each row in a table will have same fields as other rows in a same table, While in NoSQL databases, there is no fixed schema, each row can have less or additional different fields
    5. SQL databases are vertically scalable, while NoSQL are horizontally scalable
    6. SQL database eg - MySQL, Oracle, PostgreSQL, NoSQL database eg - Cassandra, MongoDB, HBase

### Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

- DDL - is a acronym for Data Definition Language. DDL statements are used to make & modify schemas of tables,databases and any objects of the database

1. CREATE 
    -  It is used to create the databse,tables,views,etc
    - eg. create the database named 'test'
    - CREATE database test;
    - eg. create table demo inside test database
    - CREATE table test.demo (id int, name varchar2(10));

2. DROP
    - It is used to delete the entire object (databse, tables, etc)
    - eg. drop the table demo
    - DROP table test.demo;
    - eg. drop the database
    - DROP database test;
    - DROP statement drops the entire mentioned object including structure of object and data in it

3. ALTER
    - It is used to modify/alter the structure of databses and objects of databses previously created
    - eg - alter table demo and add new column as address
    - ALTER table test.demo ADD address varchar2(100);

4. TRUNCATE
    - It is used to removes all the data from table
    - eg - TRUNCATE table test.demo;
    - If we truncate table it only deletes the data & not the structure of table (schema)


### Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

- DML - Data Manupulation Language. DML statements are used to manipulate/modify the actual data/rows

1. INSERT 
    - It is used to insert the data into the tables
    - consider the table demo with columns as id int, fname varchar2(10), lname varchar2(10)
    - INSERT INTO demo (id,fname,lname) VALUES (1001,'jhon','Smith'); this will insert one row into table demo

2. UPDATE
    - It is used to update values already present inside the table rows
    - eg. update the first name to 'James' where id is 1001
    - UPDATE demo SET fname='James' WHERE id=1001; this will update fname for all rows which satisfy the condition mentioned in where clause

3. DELETE
    - It is used to delete the rows from the table.
    - eg. Delete the rows where id is 1001
    - DELETE FROM demo WHERE id=1001;
    - Above sql statement will delete all rows where id is 1001

### Q4. What is DQL? Explain SELECT with an example.

- DQL - Data Query Language. Used to fethc data from tables.

1. SELECT - 
    - When we want to get data from one or more tables we can use select statements
    - SELECT will get the data from table based on conditions mentioned if any and result will be stored in temporary table.
    - We can also perform operation on the data using various in built functions - row functions, aggregate functions
    - eg. fetch all rows and all columns from table demo - SELECT * FROM demo;
    - eg. fetch fname & lname whose id is greater than 2000 - SELECT fname, lname FROM demo WHERE id > 2000;


### Q5. Explain Primary Key and Foreign Key.

- Primary Key -

    - Primary Key is used to indentify the each row uniquely.
    
- Foreign Key -
    - Foreign Key is used to link two table in relational database
