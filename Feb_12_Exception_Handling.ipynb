## 

Exception in python is a incident which occurs during execution of pthon code, which causes the python to inturpt the normal execution of programm. Exception happens when python can not handle certain commands in code

#### Exception vs Syntax errors -
1. Syntax error happens when there is a syntax error in code. While exceptions are raised when syntax of code is correct but the code gives error 

2. When syntax error happens execution of code stopped, but in case of exception execution flow of code is changed.

## 

If we do not handle the exception, python prints the error message and type of exception error occured to the standard output

```python
# eg. 
98/0
```

try & except statements are used to catch and handle exceptions. code is put inside try block which may throw exception & code is kept inside except block will handle the exception.

eg. divide by zero

```python
a=10
try:
    # this line throw error 
    res = a/0
except Exception as e:
    # we are catching the error and handling here
    print(e,"error occured bcause we can not divide number by 0.")
```

- try & else -
    - If exception is not raised in try block python will not execute the except block instead it will execute the else block 

1. As we can see below exception is raised in try, so it will execute only except block and not else.
2. In 2nd code as exception is not raised only else block is executed.

```python
try:
    print('Inside try')
    10/0
except Exception as e:
    print('Inside except block')
else:
    print('Inside else')

print('Outside of try except else blocks')
```

```python
try:
    print('Inside try')
    10/2
except Exception as e:
    print('Inside except block')
else:
    print('Inside else')

print('Outside of try except else blocks')
```

- finally -
    - finally block will be always executed irrespective of exception is raised or not
    
eg. As we can see below code inside finally block is executing no matter if exception is raised or not

```python
try:
    print('Inside try')
    # exception raised
    10/0
except Exception as e:
    print('Inside except')
finally:
    print('Inside finally')
```

```python
try:
    print('Inside try')
    #exception not raised
    10/2
except Exception as e:
    print('Inside except')
finally:
    print('Inside finally')
```

- raise -
    - raise is used to raise exception based on user logic i.e custom exceptions
    - usually such code does not raise any python defined exceptions but we want to raise exception according to our logic
    - we can define which type of exception to raise
    
eg. consider below example - where we only want to store age which is more than 18 years

```python
# storing age less than 18
age = 15

#using try except block
try:
    
    if age < 18 :
        # exceptin will be raised here
        raise Exception('Age lower than 18 not allowed')

except Exception as e:
    print(e)
```

1. Custom exceptions are the user defined exceptions based on user logic
2. To create custom exception we can inherit from "Exception" base class.
3. Custom exception allows user to have exceptions customised to their own requirements

eg. Lets take above example If we and value for age to greater than or equal to 18. Then we can define custom exception class for this logic of age.

using above age example please see below code for custom exception

```python
# creating custome exception by extending base Exception class
class ageError(Exception):
    def __init__(self):
        # defining error message by passing message to super class
        super().__init__('Age should be >= 18')

try:
    age=10
    if age<18 :
        raise ageError
except ageError as e:
    print(e)
```
