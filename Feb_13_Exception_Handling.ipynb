- ### ArithmaticError has 3 error -
    - #### ZeroDivisionError -
        - This error happens when there is attempt to divide by zero
    - #### OverflowError -
        - This error occurs when arithmetic operation exceeds the limits
    - #### FloatingPointError -
        - It is raised when there is error occures in floating point operations

```python
# ZeroDivisionError -

try:
    10/0
except ArithmeticError as e:
    print(e,e.__class__)
```

1. LookupError calss is used to catch the KeyError, IndexError as it is a base class for both of them
2. LookupError is raised when key or index can not be found for dictionary or sequence.

- KeyError - It is a subclass of LookupError and is raised when Key is not found & there is issue retrieving that key
eg. 

```python
d = {'name':'abc','contact':987654312}
try:
    #accessing the key which is not present in above dictionary
    d['email']
except KeyError as e:
    print('Key not found ',e,e.__class__)
```

- IndexError - It is a subclass of LookupError and is raised when index we are looking is not found. Like accessing index from any sequence like list which is out of bound

```python
lst = [0,1,2,3,4]
try:
    #accessing index which is not present in above list
    lst[100]
except IndexError as e:
    print(e,e.__class__)
```

- ImportError is raised when importing of module/member from module fails. Reason for this could be as below -
    1. Module which is being imported is not exist Or
    2. If we are importing member from module and that member is not part of the mentioned module

If imported module is not present then we will get ModuleNotFoundError -

```python
try:
    #importing module xyz which does not exist
    import xyz
except ImportError as e:
    print(e,e.__class__)
```
